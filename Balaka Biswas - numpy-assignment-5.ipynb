I want you to create a function which takes 2 inputs - the Mean and Standard Deviation and prints out the graph of the Normal Distribution (also known as PDF, you will find the formula in the wiki article). Assume x is given and calculate y using the pdf formula. The resultant graph would look something like this (any one of them):!

![](https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/Normal_Distribution_PDF.svg/340px-Normal_Distribution_PDF.svg.png)

```
import numpy as np
import math as m
import matplotlib as ml
import matplotlib.pyplot as plt
%matplotlib inline
ml.style.use('ggplot')

# Write function code here
def plotter(meanValue,sdValue):
  x=np.linspace(-10,10,100)
  y = []
  mn = meanValue
  sd = sdValue

  for i in x:
    p = (-((i-mn)/sd)**2)/2
    res = (1/(sd*(np.sqrt(2*np.pi))))*(m.exp(p))
    y.append(res)

  plt.plot(x,y)
  plt.show()


# Test
plotter(0,np.sqrt(0.2))
```

```
# You have write a function which takes 2 numpy array as input and returns "Hoga" or "Nahi Hoga" 
# depending on if broadcasting is possible on those two arrays
```

```
# Write your logic here

# Conditions if two arrays are broadcastable :
# 1. Arrays have excatly same shape
# 2. Same number of dimensions, length is common or 1
# 3. Arrays can be prepended with 1 (if smaller)
def hoga_ki_nahi(arr1,arr2):
  # Using zip function for mapping (point 3)
  if all((m==n) or (m==1) or (n==1) for m, n in zip(arr1.shape[::-1],arr2.shape[::-1])):
    return "Hoga"
  else:
    return "Nahi Hoga"
```

```
# Test your function here

ar1 = np.zeros((3,1,5))
ar2 = np.zeros((4,5))
ar3 = np.zeros((5,3))

print("Can ar1 and ar2 be broadcasted ? Ans : ", hoga_ki_nahi(ar1,ar2))
print("Can ar1 and ar3 be broadcasted ? Ans : ", hoga_ki_nahi(ar1,ar3))
```
