**1. IMPLEMENTING NUMPY.LINSPACE() FROM SCRATCH**

```
def linspace(start,stop,num=50,endpoint=True):            # num is 50 by default, endpoint is True by default
  new_l = []
  if endpoint:
    interval = (stop - start)/(num - 1)
    # If ending point is the number defined as stop, then the number of intervals we'll calculate decrease by one, as one interval has been fixed/one position has been captured already
  else:
    interval = (stop - start)/num
  i = 1
  while i <= num:
    if i == 1:
      new_l.append(float(start))
    else:
      new_l.append(float(start + (interval*(i-1))))
    i = i + 1
  return new_l             # Return Python list
```
